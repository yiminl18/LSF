```python
import os

folder = "/Users/yiminglin/Documents/Codebase/LSF/data/provenance_results/Affiliate_Agreements"
all_paths = []

for root, dirs, files in os.walk(folder):
    for file in files:
        all_paths.append(os.path.join(root, file))

print(all_paths)
```

```python
import json 

questions = []

for path in all_paths:
    print('path:', path) 
    
    data = json.load(open(path, 'r'))
    i = 0
    for doc in data:
        print(i)
        i += 1 
        question = doc['question_text']
        true_answer = doc['true_answer']
        llm_global_answer = doc['llm_global_answer']
        doc_path = doc['doc_path']
        print('question:', question)
        print('true_answer:', true_answer)
        print('llm_global_answer:', llm_global_answer)
        print('doc_path:', doc_path)
        provenance_nodes = doc['provenance_nodes']
        print('len(provenance_nodes)', len(provenance_nodes))
        print('provenance info:')
        for node in provenance_nodes:
            node_summary = node['node_summary']
            header_text = node['header_text']
            page_no = node['page_no']
            total_pages = node['total_pages']
            tree_node = node['tree_node']
            tree_header_num = node['tree_header_num']
            print('node_summary', node_summary)
            print('header_text', header_text)
            print('page_no', page_no)
            print('total_pages', total_pages)
            print('tree_node', tree_node)
            print('tree_header_num', tree_header_num)
            break 
    print('--------------------------------')       
     

```
